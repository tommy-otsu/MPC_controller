```python
from numpy import sin,cos
from scipy.integrate import solve_ivp,odeint # 積分
from matplotlib.animation import FuncAnimation
import matplotlib.pyplot as plt
import numpy as np
```

```python
G = 9.8
L = 1.0
```

```python
def state_eq(state,t,u):
    dxdt = np.zeros_like(state)
    f = np.array([state[1], (G/L)*sin(state[0])]) 
    g = np.array([0,1])
    dxdt = f + g*u
    return dxdt

# 時刻
t_span = [0,20]
dt = 0.05
t = np.arange(t_span[0], t_span[1],dt)

# 初期状態
th1 = 0.1
dth1 = 0
state = np.array([th1,dth1])

# 積分計算：solve_ivp(微分方程式，時間区間，初期状態，)
#sol = odeint(state_eq,state,t,"パラメータがあれば代入") 
# 制御入力をパラメータとして代入
sol = odeint(state_eq,state,t,args=(7,))
theta = sol[:,0]
print(np.shape(sol))

plt.plot(t,theta,label="test")
plt.legend()
plt.xlabel("time [s]")
plt.ylabel(r"${\theta}$[rad]") # rを先頭につけることでlatex仕様に
plt.show
```

```python
x = L * sin(theta)       # x = Lsin(theta)
y = L * cos(theta)      # y = Lcos(theta)


# line, = ax.plot([], [], 'o-', linewidth=2) # このlineに次々と座標を代入して描画

def animate_pendulum(t,x,y):
    fig, ax = plt.subplots()
    for i in np.arange(len(t)):
        line, = ax.plot(x[i],y[i],"o", color='black')
    # グラフを描画する
        line, = ax.plot([0,x[i]], [0,y[i]], color='black')
    #plt.plot([ax, hx],[ay, hy], color='green', linestyle='--', zorder=0)
    # 次の描画まで0.01秒待つ
        plt.pause(0.01)
        ax.set_aspect('equal')
        ax.set_xlim(-1.2*L,1.2*L)
        ax.set_ylim(-1.2*L,1.2*L)
        plt.xlabel("x")
        plt.ylabel("y")
    # グラフをクリア
        line.remove()

#ani = FuncAnimation(fig, animate, frames=np.arange(0, len(t)), interval=25, blit=True)

#ax.set_xlim(-L,L)
#ax.set_ylim(-L,L)
#a
#ax.grid()
#plt.show()

#ani.save('pendulum.gif', writer='pillow', fps=15)
```
